Task.4 :- Suppose you have a DataFrame of Instagram usernames where some entries have typos (like 'insta_queen', 'insta-queen', 'instaqueen'). Use the replace() function to standardize all these variants to 'instaqueen'.


```python
df4 = pd.DataFrame({
    "username": [
        "insta_queen",
        "insta-queen",
        "instaqueen",
        "travel_girl",
        "insta_queen",
        "foodlover",
        "insta-queen"
    ]
})
df4
```

```python
df4["username"] = df4["username"].replace({"insta_queen" : "instaqueen","insta-queen":"instaqueen"})
df4
```

Task.5 :- You have a DataFrame column for payment status from a Paytm-like app with mixed values: 'Yes', 'yes', 'Y', 'No', 'no', 'N', and some with extra spaces. Write code to unify this column so all paid statuses become 1 and all unpaid statuses become 0, trimming whitespace and fixing capitalization where needed.<br><br><em><strong>Hint:</strong> Use str.strip(), str.lower(), and map/replace methods in pandas.</em>

```python
df4 = pd.DataFrame({
    "values" : ['Yes', 'yes', 'Y'," y ", 'No', 'no', 'N',   " n ","   yes "    ]
})
df4
```

```python
df4["values"] = df4["values"].str.strip().str.lower().replace({"yes":"1","y":"1","no":"0","n":"0"})
df4
```
