Task.2 :- Load a CSV file containing order dates from a Flipkart-style order history (column: 'order_date', format: 'YYYY-MM-DD HH:MM:SS'). Extract the year, month, and weekday for each order and add them as new columns in the DataFrame.


```python
df2 = pd.read_csv("flipcartstyle.csv")
df2
```

```python
df2["order_date"] = pd.to_datetime(df2["order_date"])
```

```python
df2["Year"] = df2["order_date"].dt.year
df2["Month"] = df2["order_date"].dt.month
df2["Date"] = df2["order_date"].dt.day
df2["Day"] = df2["order_date"].dt.weekday
df2
```
