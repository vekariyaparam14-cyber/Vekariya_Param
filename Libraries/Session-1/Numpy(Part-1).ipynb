Task.2:- Create a NumPy array called followers using np.array() that stores the follower counts for 5 Instagram influencers: [1200, 15000, 67000, 340000, 1250000]. Print the array, its shape, number of dimensions, and data type.


```python
followers = np.array([1200, 15000, 67000, 340000, 1250000])

print("array :",followers)
print("Shape :",np.shape(followers))
print("Dimensions :",np.ndim(followers))
print("datatype:",followers.dtype)
```
