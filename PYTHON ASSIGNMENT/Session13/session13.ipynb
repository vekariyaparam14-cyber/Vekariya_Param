### Task.1 :- Write a recursive function in Python called print_playlist_songs(songs) that takes a list of song names (like a Spotify playlist) and prints each song name one by one using recursion.


```python
a = ['Shape Of You', 'Blinding Lights', 'Levitating', 'Senorita']

def playlist_songs(song):
    if len(song) == 0:
        return 0
    print(song[0:])
    
playlist_songs(a)
```

```python

x = 'global'
def outer():
    x = 'outer'
    print("Inside:",x)

outer()
x
```

### Task.4 :- Build a recursive function format_number_short(n) that takes a number (like a follower count on Instagram or YouTube) and returns it as a string in short format: 1500 as '1.5K', 1200000 as '1.2M', 500 as '500'.

```python
def format_followers(n):
    if n >= 1000000:
        return f"{n / 1000000:.2f}M"
    elif n >= 1000:
        return f"{n / 1000:.2f}K"
    else:
        return str(n)

print(format_followers(500))
print(format_followers(1500))
print(format_followers(25000))
print(format_followers(1200000))
```
