```python

x = 'global'
def outer():
    x = 'outer'
    print("Inside:",x)

outer()
x
```

```python
x = 'global'
def inner():
    global x
    x = 'inner'
    
inner()
print('Inside:', x)
x
```

```python
"""
FROM ABOVE 
IN def outer() ---> x is local because the value of x(outer) only use in inside def function not in outside of def function.
IN def inner() ---> x is global becauce the value of x(inner) in outside def functino is same as inside def function.
"""
print("")
```

### Task.4 :-

```python
def format_followers(n):
    if n >= 1000000:
        return f"{n / 1000000:.2f}M"
    elif n >= 1000:
        return f"{n / 1000:.2f}K"
    else:
        return str(n)

print(format_followers(500))
print(format_followers(1500))
print(format_followers(25000))
print(format_followers(1200000))
```
