### Task.1 :- Create a Python class called Playlist with a private attribute _songs (a list) and a public method add_song(song) to add a song title to the playlist. Print the playlist after adding 3 songs.


```python
class playlist:
    def __init__(self, song,):
        self.__song = song
        
    def add_song(self, add_song):
        self.__add_song = self.__song.extend(add_song)
        
    def show_playlist(self):
        return self.__song
```

### Task.2 :-Build a Product class for a Flipkart-style app with a private attribute _price. Implement get_price() and set_price() methods to access and update the price. Demonstrate setting and getting the price for a product object.

```python
class flipcart_style:
    def __init__(self, price):
        self.__price = price
        
    def set_price(self, price):
        self.__price = 0 if price < 0 else price
    
    def get_price(self):
        return self.__price
```

```python
Flipcart = flipcart_style(650)
```

```python
Flipcart.get_price()
```

```python
Flipcart.set_price(550)
```

```python
Flipcart.get_price()
```

### Task.4 :- Design a simple abstract class PaymentMethod with an abstract method pay(amount). Then, create two subclasses: Paytm and PhonePe, each implementing pay(amount) to print a different message. Instantiate both and call their pay methods with any amount.

```python
from abc import ABC, abstractmethod

class payment_method(ABC):
    
    @abstractmethod
    def pay_amount(self,amount):
        pass
```

```python
class Paytm(payment_method):
    
    def pay_amount(self, amount):
        return f"You Paid Rs.{amount}! by Paytm."
```

```python
class Phonepa(payment_method):
    
    def pay_amount(self, amount):
        return f"You Paid Rs.{amount}! by Phonepa"
```

```python
a = Paytm()
b = Phonepa()
```

```python
a.pay_amount(5000)
```

```python
b.pay_amount(6000)
```
