```python
def calc_final_price(price, dis):   
    total_dis = price*dis/100
    final_price= price - total_dis
    return final_price

calc_final_price(1000,50)
```

### Task.3 :-

```python
def format_price(price, currency="INR"):
    if currency == "INR":
        return f'₹{price}'
    else:
        return f'${price}'
    
print(format_price(500))
format_price(500,"USD")
```

### Task.4 :-

```python
def apply_coupon(price, coupon_code = "ZOMATO10"):
    if coupon_code == "ZOMATO10":
        total_dis = price*10/100
        final_price= price - total_dis
        return final_price
    else:
        return price

print(apply_coupon(1000))
apply_coupon(1000,"NONE")
```
