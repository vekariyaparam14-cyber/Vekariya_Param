### Task.1 :- Write a Python function called get_song_duration_per_minute that divides the total duration of a Spotify playlist (in minutes) by the number of songs, and handles the case where the number of songs is zero using try, except, and finally blocks.

```python
try:
    def average_duration(duration_time,number_of_song):
        return duration_time/number_of_song
    print(average_duration(120,0))
    
except ZeroDivisionError as e:
    print("The Error is: ",e)
finally:
    print("Block")
```

### Task.2 :- Build a Flipkart-style price-per-item calculator: take total cart amount and item count as input, perform division, and use try-except to catch and display a user-friendly message if the item count is zero.


```python
try:
    cart_amount = int(input("Enter your Total Cart amount:"))
    no_of_item = int(input("Enter your no. of item in cart:"))
    print(cart_amount)
    print(no_of_item)
    price_per_item = cart_amount/no_of_item
    
except (ValueError,ZeroDivisionError) as e:
    print("The error is:",e)

print("Price per item:",price_per_item)
```

### Task.5 :- Write a function called safe_divide_for_zomato that takes two numbers (bill amount and number of people), uses try, except, else, and finally to divide the bill and print the result, print a custom error if division by zero, and always print 'Split calculation done' at the end.
