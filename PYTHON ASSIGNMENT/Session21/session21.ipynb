### Task.1 :- Create a Python class called Product with attributes name and price, and a method get_discounted_price() that returns the price after applying a 10% discount.


```python
class product:
    
    def __init__(self, name, price):
        self.name = name
        self.price = price
        
    def get_details(self):
        return f"Your Product is {self.name}" ,f"You will get 10% discount.",  f"Your Total Price is {self.price - (10/100*self.price)}!"
```

```python
Product = product("Laptop",50000)
```

```python
Product.get_details()
```

### Task.2 :- Now create a subclass called Electronics that inherits from Product and overrides the get_discounted_price() method to apply a 20% discount instead of 10%.


```python
class Electronics(product):
    def __init__(self, name, price):
        self.name = name
        self.price = price
        
    def get_details(self):
        return f"Your Product is {self.name}" ,f"You will get 20% discount.",  f"Your Total Price is {self.price - (20/100*self.price)}!"
```

```python
electric = Electronics("TV", 60000)
```

```python
electric.get_details()
```

### Task.4 :- Build a simple Ticket class for a movie booking app with a method get_final_price(). Then, create a subclass PremiumTicket that overrides get_final_price() to add a 50 rupee convenience fee. Show both in action by creating objects and printing their final prices.<br><br><em><strong>Hint:</strong> Use super() in PremiumTicket to reuse the parent method and add the extra fee.</em>


```python
class ticket:
    
    def t_price(self, price):
        self.price = price
        
    def get_final_price(self):
        return f"Your Final Ticket Price: {self.price}"
```

```python
class premiun_ticket(ticket):
    
    def t_price(self, price):
        self.price = price + 50
        
    def get_final_price(self):
        return super().get_final_price()
```

```python
a = ticket()
a.t_price(6000)
```

```python
a.get_final_price()
```

```python
p = premiun_ticket()
p.t_price(6000)
```

```python
p.get_final_price()
```
