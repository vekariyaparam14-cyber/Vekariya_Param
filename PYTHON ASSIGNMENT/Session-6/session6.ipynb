### Task.3 :- Given a dictionary called food_prices with 5 Zomato food items as keys and their prices as values, write code to display all items that cost more than ₹200.


```python
zomato = {
    "7_Cheesy_Pizza" : 150,
    "Margerita_Pizza" : 100,
    "Panner_Tadka_Pizza" : 250,
    "9_Cheesy_Pizza" : 300,
    "Italiya_Pizza_Hut" : 350
}

print(zomato)
print("\n")
print("Item cost more then 200:-\n")
for item, price in zomato.items():
    if price > 200:
        print(item, "-" + str(price))
        
```
