### Task.2 :-

```python
daily_steps = [5000,6500,7000,10000,10500,12000,17000]
days = ["Monday","Tuesday","Wednesday","Thursday","Friday","Saturday","Sunday"]

i = 0
while i < len(daily_steps):
    if daily_steps[i] > 10000:
        print("First day crossing 10,000 steps:", days[i])
        print("Steps:", daily_steps[i])
        break
    i += 1
```
