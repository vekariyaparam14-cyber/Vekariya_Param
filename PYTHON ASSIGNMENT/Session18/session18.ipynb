```python
import re
text = "hello, please connect with me with +91-7990757228 and +91-8000733320"
c = re.findall(r"\+91-\d{10}",text)
c
```

```python
date = "hello today's date is 14/09/2025"

if re.search(r"[1-32]\d+/[01-12]\d+/\d{4}",date):
    print("True")
else:
    print("False")
    
d = re.findall(r"[1-32]\d+/[01-12]\d+/\d{4}",date)
d
```

```python
prices = "Laptop: Rs. 29999 , Mouse: Rs. 599 , Keyboard: Rs. 1500 , Headphones: Rs. 2499"

d = re.findall(r"\d+",prices)
d = [int(i) for i in d]

total = sum(d)
print("Prices:",d)
print("Total price:",total)
```
