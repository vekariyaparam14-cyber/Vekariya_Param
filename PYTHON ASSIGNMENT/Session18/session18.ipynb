### Task.1 :-  Use re.findall() to extract all valid phone numbers from a given string in the format '+91-XXXXXXXXXX' (e.g., '+91-9876543210'). Print the list of found numbers.


```python
import re
text = "hello, please connect with me with +91-7990757228 and +91-8000733320"
c = re.findall(r"\+91-\d{10}",text)
c
```

### Task.2 :- Write a Python function using re.search() that checks if a string contains a valid date in the format 'DD/MM/YYYY'. The function should return True if a date is found, otherwise False.


```python
date = "hello today's date is 14/09/2025"

if re.search(r"[1-32]\d+/[01-12]\d+/\d{4}",date):
    print("True")
else:
    print("False")
    
d = re.findall(r"[1-32]\d+/[01-12]\d+/\d{4}",date)
d
```

### Task.3 :- Given a block of text containing multiple prices (like 'Rs. 299', 'Rs. 1500', etc.), use re.findall() to extract all the numeric price values as integers and print their sum.

Hint: Look for patterns like 'Rs. ' followed by one or more digits.

```python
prices = "Laptop: Rs. 29999 , Mouse: Rs. 599 , Keyboard: Rs. 1500 , Headphones: Rs. 2499"

d = re.findall(r"\d+",prices)
d = [int(i) for i in d]

total = sum(d)
print("Prices:",d)
print("Total price:",total)
```

### Task.4 :- Use re.sub() to replace all email addresses in a string with '[hidden email]' and print the modified string.

Constraint: Do not use any external libraries except re.

```python
string = "hello, please connect with me with abc@gmail.com and yahoo@gmail.com"
email = re.sub(r'[a-zA-Z0-9.%+-]+@[a-zA-Z0-9.-]+\.[a-zA-Z]{2,}',"[Hidden email]", string)

email
```
