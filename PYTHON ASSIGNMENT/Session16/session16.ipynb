### Task.2 :-

```python
list2 = ['Shape Of You', 'Blinding Lights', 'Levitating', 'Senorita']
def playlist_generator(n = len(list2)):
    for i in range(n):
        yield i
        
gen = playlist_generator()

for i in gen:
    print(i,list2[i])
```
