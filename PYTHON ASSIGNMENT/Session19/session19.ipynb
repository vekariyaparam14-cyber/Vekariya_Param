##### Task.1 :- Define a Python class called Song with attributes title, artist, and duration (in seconds). Create an object for your favorite song and print its details.

##### Task.2 :- Add a method play_preview(self) to the Song class that prints 'Playing 30-second preview of [title] by [artist]'. Call this method using the object you created.


```python
class song:
    def __init__(self, title, artist, duration_in_sec):
        self.title = title
        self.artist = artist
        self.duration_in_sec = duration_in_sec
        
    def song_details(self):
        return self.title, self.artist, self.duration_in_sec
    
    def play_review(self):
        return f"Playing 30-second preview of {self.title} by {self.artist}"
```

```python
Param = song("Hanuman Chalisa","Aditya Gadhvi",600)
```

```python
Param.song_details()
```

```python
Param.play_review()
```

### Task.3 :- Create a class called FoodOrder with attributes: restaurant_name, items (list), and total_price. Write an __init__() constructor to initialize these, then create an object representing your last Zomato or Swiggy order and print its details.

```python
class FoodOrder:
    def __init__(self,restaurant_name, items, total_price):
        self.restaurant_name = restaurant_name
        self.items = items
        self.total_price = total_price
        
    def order_details(self):
        return self.restaurant_name, self.items, self.total_price
```

```python
Zomato = FoodOrder("La Milano'Z",["Margerita Pizza","Italian Pizza","French Authentic Pizza"],999)
```

```python
Zomato.order_details()
```
