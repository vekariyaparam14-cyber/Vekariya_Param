```python
flipcart_cart = int(input("Enter your Cart Amount:"))
print("Your Cart amount is:",flipcart_cart)

if flipcart_cart >= 2000:
    print("You will get 10% Discount")
    Total = flipcart_cart - (flipcart_cart * 10/100)
    print("Your Total Amount is: ",Total)
elif flipcart_cart >= 1000:
    print("You will get 5% Discount")
    Total = flipcart_cart - (flipcart_cart * 5/100)
    print("Your Total Amount is: ",Total)
else:
    print("You Will NOT GET DISCOUNT:")
    Total = flipcart_cart
    print("Your Total Amount is: ",Total)
```
