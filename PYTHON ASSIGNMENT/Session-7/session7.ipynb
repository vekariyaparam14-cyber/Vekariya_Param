### Task.1 :- Write a Python script that takes your current Spotify listening time in minutes and checks if it is above 120 minutes; if yes, print 'You are a true music fan!', otherwise print 'Keep listening!'.


```python
a = int(input("Enter your Spotify music listening minutes:"))

print("My listening time(min.): ",a)

if a >= 120:
    print("You are a true music fan!")
else:
    print("Keep listening!")
```

### Task.2 :- Create a Python program that asks the user to enter their Zomato order amount and checks if it is above 300; if yes, print 'Eligible for free delivery', else print 'Delivery charges apply'.

### Task.3 :- Build a Python script that takes your Flipkart cart total and applies the following logic: if total > 2000, print 'You get a 10% discount'; elif total > 1000, print 'You get a 5% discount'; else print 'No discount available'.


```python
flipcart_cart = int(input("Enter your Cart Amount:"))
print("Your Cart amount is:",flipcart_cart)

if flipcart_cart >= 2000:
    print("You will get 10% Discount")
    Total = flipcart_cart - (flipcart_cart * 10/100)
    print("Your Total Amount is: ",Total)
elif flipcart_cart >= 1000:
    print("You will get 5% Discount")
    Total = flipcart_cart - (flipcart_cart * 5/100)
    print("Your Total Amount is: ",Total)
else:
    print("You Will NOT GET DISCOUNT:")
    Total = flipcart_cart
    print("Your Total Amount is: ",Total)
```

### Task.4:- Write a Python program that asks the user to enter their IPL fantasy team points and uses nested if-else statements to print: 'Champion' if points > 800, 'Top Performer' if points between 500 and 800, 'Keep Trying' otherwise.

Hint: Use nested if-else blocks to check the ranges.

```python
fantasy_score = int(input("Enter Your FANTASY TEAM SCORE:"))
print("Your FANTASY SCORE: ",fantasy_score)

if fantasy_score > 800:
    print("Champion!")
elif fantasy_score > 500:
    print("Top Performer!")
else:
    print("Keep Trying!")
```
